```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.random.randint(1000, size=50)
b = np.random.randint(1000, size=50)
fig, ax = plt.subplots()
col = np.where((a+b)%2 == 0,'k','r')
ax.scatter(a,b,c=col)
ax.set_title('Even Scatter Plot')
ax.set_xlabel('horizontal axis')
ax.set_ylabel('vertical axis')
```

```python
x = np.array(range(0,5))  
y = eval('3*x + 0.5')
fig, ax = plt.subplots()
ax.plot(x, y,marker='>',label = '3x',color='r')  
x2 = np.array(range(0,5))  
y2 = eval('5*(x**(.5))')
ax.plot(x2, y2,linestyle='--',marker='*',label = 'squareroot',color='b')  
ax.legend()

```

```python
x = np.array(range(0,5))  
y = eval('3*x + 0.5')
fig, ax = plt.subplots()
ax.plot(x, y,marker='>',label = '3x',color='r')  
x2 = np.array(range(0,5))  
y2 = eval('5*(x**(.5))')
ax.plot(x2, y2,linestyle='--',marker='*',label = 'squareroot',color='b')  
ax.legend()
ax.set_xscale('log')
```

```python
barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
fig, ax = plt.subplots()
ax.bar(np.arange(len(barheights)), barheights)
x_pos = np.arange(len(barheights))
ax.set_xticks(x_pos)
ax.set_xticklabels(barlabels, rotation=0)
```
